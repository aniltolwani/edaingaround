```python
import gzip
import json
from pathlib import Path
from datetime import datetime, timedelta
import random

def create_test_data():
    """Create the complete test directory structure with sample data."""
    
    base_dir = Path("project_data")
    
    # Sample code responses
    code_samples = {
        "py": [
            "def fibonacci(n):\n    if n <= 1:\n        return n\n    return fibonacci(n-1) + fibonacci(n-2)",
            "import pandas as pd\nimport numpy as np\n\ndf = pd.DataFrame({'A': np.random.randn(100), 'B': np.random.randn(100)})\nprint(df.describe())",
            "class DataProcessor:\n    def __init__(self, data):\n        self.data = data\n    \n    def process(self):\n        return [x * 2 for x in self.data]",
            "print('hello')",  # Short one
            "x = 5",  # Another short one
        ],
        "js": [
            "const fetchUserData = async (userId) => {\n    const response = await fetch(`/api/users/${userId}`);\n    const data = await response.json();\n    return data;\n}",
            "function calculateSum(arr) {\n    return arr.reduce((acc, val) => acc + val, 0);\n}",
            "console.log('test');",  # Short one
            "const numbers = [1, 2, 3, 4, 5].map(x => x * 2).filter(x => x > 5).reduce((a, b) => a + b, 0);",
        ],
        "java": [
            "public class Main {\n    public static void main(String[] args) {\n        System.out.println(\"Hello World\");\n    }\n}",
            "import java.util.*;",  # Short one
        ],
        "rust": [
            "fn main() {\n    println!(\"Hello, world!\");\n}",
        ]
    }
    
    prompts = [
        "write a function", "create a class", "implement algorithm",
        "data processing", "async operation", "error handling",
        "parse JSON", "handle API call", "transform data"
    ]
    
    # Create directory structure
    (base_dir / "2024" / "jan").mkdir(parents=True, exist_ok=True)
    (base_dir / "2024" / "feb").mkdir(parents=True, exist_ok=True)
    (base_dir / "2025").mkdir(parents=True, exist_ok=True)
    
    print("Creating test files...")
    
    # 1. Create 2024/jan/code_analysis.jsonl.gz (compressed JSONL)
    print("  Creating 2024/jan/code_analysis.jsonl.gz...")
    with gzip.open(base_dir / "2024" / "jan" / "code_analysis.jsonl.gz", "wt") as f:
        base_time = datetime(2024, 1, 15, 10, 30, 0)
        for i in range(50):  # 50 records
            lang = random.choice(["py", "js", "java", "rust"])
            record = {
                "id": f"jan_{i:03d}",
                "lang": lang,
                "prompt": random.choice(prompts),
                "response": random.choice(code_samples[lang]),
                "timestamp": (base_time + timedelta(minutes=i)).isoformat()
            }
            f.write(json.dumps(record) + "\n")
    
    # 2. Create 2024/jan/metrics.json (regular JSON, not compressed)
    print("  Creating 2024/jan/metrics.json...")
    metrics_data = {
        "month": "january",
        "year": 2024,
        "total_completions": 50,
        "languages": {"py": 20, "js": 15, "java": 10, "rust": 5}
    }
    with open(base_dir / "2024" / "jan" / "metrics.json", "w") as f:
        json.dump(metrics_data, f, indent=2)
    
    # 3. Create 2024/feb/code_analysis.jsonl.gz (compressed JSONL)
    print("  Creating 2024/feb/code_analysis.jsonl.gz...")
    with gzip.open(base_dir / "2024" / "feb" / "code_analysis.jsonl.gz", "wt") as f:
        base_time = datetime(2024, 2, 1, 9, 0, 0)
        for i in range(30):
            lang = random.choice(["py", "js", "java"])
            record = {
                "id": f"feb_{i:03d}",
                "lang": lang,
                "prompt": random.choice(prompts),
                "response": random.choice(code_samples[lang]),
                "timestamp": (base_time + timedelta(hours=i)).isoformat()
            }
            f.write(json.dumps(record) + "\n")
    
    # 4. Create 2024/feb/code_analysis_v2.jsonl (uncompressed JSONL)
    print("  Creating 2024/feb/code_analysis_v2.jsonl...")
    with open(base_dir / "2024" / "feb" / "code_analysis_v2.jsonl", "w") as f:
        base_time = datetime(2024, 2, 15, 14, 0, 0)
        for i in range(25):
            lang = random.choice(["py", "js"])
            record = {
                "id": f"feb_v2_{i:03d}",
                "lang": lang,
                "prompt": random.choice(prompts),
                "response": random.choice(code_samples[lang]),
                "timestamp": (base_time + timedelta(hours=i*2)).isoformat(),
                "version": "2.0"
            }
            f.write(json.dumps(record) + "\n")
    
    # 5. Create 2025/code_full.json.gz (regular JSON, compressed)
    print("  Creating 2025/code_full.json.gz...")
    results = []
    base_time = datetime(2025, 1, 1, 0, 0, 0)
    for i in range(40):
        lang = random.choice(["py", "js", "java", "rust"])
        results.append({
            "id": f"2025_{i:03d}",
            "lang": lang,
            "prompt": random.choice(prompts),
            "response": random.choice(code_samples[lang]),
            "timestamp": (base_time + timedelta(minutes=i*10)).isoformat()
        })
    
    full_data = {
        "results": results,
        "metadata": {
            "version": "2.0",
            "created": datetime.now().isoformat(),
            "total_count": len(results)
        }
    }
    
    with gzip.open(base_dir / "2025" / "code_full.json.gz", "wt") as f:
        json.dump(full_data, f, indent=2)
    
    # 6. Create metadata.json (root level)
    print("  Creating metadata.json...")
    metadata = {
        "project": "code_analysis",
        "created": datetime.now().isoformat(),
        "description": "Test data for JSON processing pipeline",
        "structure": {
            "2024": ["jan", "feb"],
            "2025": ["full_year"]
        }
    }
    with open(base_dir / "metadata.json", "w") as f:
        json.dump(metadata, f, indent=2)
    
    # 7. Create a "trap" file - JSONL file with .json extension (mislabeled)
    print("  Creating 2024/feb/mislabeled.json (actually JSONL)...")
    with open(base_dir / "2024" / "feb" / "mislabeled.json", "w") as f:
        for i in range(10):
            record = {
                "id": f"trap_{i:03d}",
                "lang": "py",
                "response": random.choice(code_samples["py"]),
                "timestamp": datetime.now().isoformat()
            }
            f.write(json.dumps(record) + "\n")
    
    print("\nTest directory structure created!")
    print("\nFiles created:")
    for file_path in base_dir.rglob("*"):
        if file_path.is_file():
            size = file_path.stat().st_size
            print(f"  {file_path.relative_to(base_dir)} ({size} bytes)")
    
    # Print some statistics
    print("\nData statistics:")
    print(f"  Total JSONL records: ~145")
    print(f"  Total JSON array records: 40")
    print(f"  Languages included: Python, JavaScript, Java, Rust")
    print(f"  Mix of compressed and uncompressed files")
    print(f"  One mislabeled file (mislabeled.json is actually JSONL)")

if __name__ == "__main__":
    create_test_data()
    print("\nNow you can run your processing script on 'project_data/' directory!")
```
